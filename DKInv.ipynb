# 8. We might be interested in ...

```python
# finding the top 5 performers in each asset_class

top_performances = (
    inv.groupby("asset_class", group_keys=False)
    .apply(lambda x: x.nlargest(5, "3y_ann_perf"))
top_performances
```

```python
# might be better to vizualise the the top 5 performances in each asset class. note we only have 2 fund in alternatives and real estate

plt.figure(figsize=(12, 6))
sns.barplot(x='fund_name', y='3y_ann_perf', data=top_performances, hue='fund_name', errorbar=None, palette='viridis', legend=False)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right", fontsize=10)  # Rotate 45 degrees, align right, adjust font size

# Add labels and title
plt.title('Top 5 Performances in Each Asset Class', fontsize=16)
plt.xlabel('fund_name', fontsize=12)
plt.ylabel('3Y Performance (%)', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# we coould look at ann_cost in relation to the 3y_ann_perf

plt.figure(figsize=(15,8))

sns.scatterplot(x="ann_cost", y="3y_ann_perf", data = top_performances , color="skyblue", edgecolor="black", s=100, lw=2)

# Add fund_name labels to each dot using iterrows()
for _, row in top_performances.iterrows():
        plt.text(
        x=row["ann_cost"], 
        y=row["3y_ann_perf"] + 0.2,  # Slight vertical offset
        s=row["fund_name"],
        fontsize=10,
        color="black",
        ha="center"
    )

# Add titles and labels
plt.title("Annual Cost vs 3Y Annual Performance", fontsize=16)
plt.xlabel("Annual Cost (%)", fontsize=12)
plt.ylabel("3Y Annual Performance (%)", fontsize=12)

plt.tight_layout()
plt.show()

# seems like we "get" much performance relative to cost with Storebrands and Sparindex funds...
```

```python
# as we can see by the sharpe_ratio as well (though sorting only by sharpe_ratio would leave out some, including Storebrand, due to NaN)

top_performances.sort_values("sharpe_ratio", ascending=False)
```

```python
# Remove any rows with NaN values in 'AUM_DKK' or 'fund_name'
inv = inv.dropna(subset=["AUM_DKK", "fund_name"])

# Limit to the top 5 funds based on 'AUM_DKK'
top_funds = inv.nlargest(5, "AUM_DKK")

# Create a category for the remaining funds
other_funds = inv.loc[~inv["fund_name"].isin(top_funds["fund_name"])]
other_sum = other_funds["AUM_DKK"].sum()

# Create a DataFrame for "Other" funds
other_df = pd.DataFrame({"fund_name": ["Other"], "AUM_DKK": [other_sum]})

# Concatenate the top funds with the "Other" category
top_funds = pd.concat([top_funds, other_df], ignore_index=True)

# Apply a Seaborn theme
sns.set_theme(style="whitegrid")

# Create the pie chart
plt.figure(figsize=(10, 10))
plt.pie(
    top_funds["AUM_DKK"],
    labels=top_funds["fund_name"],
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("pastel"),
    wedgeprops={'edgecolor': 'black'}
)

# Add a title and labels
plt.title("AUM_DKK", fontsize=16)
plt.xlabel("3-Year Annual Performance")
plt.ylabel("Frequency")

# Show the pie chart with a legend for better clarity
plt.legend(top_funds["fund_name"], title="Funds", loc="center left", bbox_to_anchor=(1, 0.5))
plt.show()
```
